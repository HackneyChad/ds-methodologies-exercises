```python
# Combined fuel economy is a weighted average of City and Highway MPG values that is calculated
# by weighting the City value by 55% and the Highway value by 45%.
# (hwy * .45) + (cty * .55)
# per https://www.fueleconomy.gov/feg/label/learn-more-gasoline-label.shtml


# This defines what the new column does, or holds
combined_fuel_econ = (df_gas_mileage.hwy * .45) + (df_gas_mileage.cty * .55)

# this creates the new column in the data set
df_gas_mileage['combined_fuel_econ'] = combined_fuel_econ

# this calls the column to print
df_gas_mileage


#SEE ZACH'S EXAMPLE ON HOW TO STRING ALL THIS TOGETHER USING THE df.assign function
# GREAT stuff!

                          
#store each intermediate step in a variable
# or backslashes to continue experessoin over multiple lines



# SEE ZACH'S EXAMPLE ON GITHUB
# HE WORKS THIS OUT PIECE BY PIECE AS SEPARATE VARIABLES


# df_gas_mileage[['manufacturer','hwy','cty']].groupby('manufacturer').mean().sort_values(by='hwy')

# df_gas_mileage[['manufacturer','hwy','cty']]\
#     .groupby('manufacturer')\
#     .mean()\
#     .sort_values(by='hwy', ascending=False)

# or you can wrap this all in parens to tell it how to handle this, but remove the backslashes:
(df_gas_mileage[['manufacturer','hwy','cty']]
    .groupby('manufacturer')
    .mean()
    .sort_values(by='hwy', ascending=False))
```

How many animals are hoppers that are above the median speed? What percentage is this?

```python
hoppers = df_fuzzy[df_fuzzy.hoppers]
median_speed = df_fuzzy.speed.median()

n_hoppers = hoppers[hoppers.speed > median_speed].shape[0]
percent_hoppers = n_hoppers / df_fuzzy.shape[0]
print('{} are hoppers, which is {:.2%} of the total'.format(n_hoppers, percent_hoppers))
```
