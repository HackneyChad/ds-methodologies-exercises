```python
def describe_grade(number_grade: int) -> str:
    if number_grade > 90:
        return 'Great!'
    elif number_grade > 70:
        return 'Ok'
    else:
        return 'Uh oh...'
    
df.MATH_GRADE.apply(describe_grade)
# this is like a map filter reduce operation, in python
```

```python
def curve_grade(number_grade: int):   # don't necess
    new_grade = number_grade + 10
    if new_grade > 100:
        return 100
    else:
        return new_grade
    
df.MATH_GRADE.apply(curve_grade)

# this passes the curve_grade function in the top of this cell, INTO the "math_grade.apply() function" from above cell
# "apply" function is a function that accepts a function as its argument.
# it accepts a function or returns a function
# this is a good use case for using lambdas
```

```python
# use of lambda in the 'apply' function
df.READ_GRADE.apply(lambda grade: 'passing' if grade >= 70 else 'failing')
```

```python
df['curved_math_grade'] = df.MATH_GRADE.apply(curve_grade)
# this creates a new field in the data set now with the curve of the existing math grade, now call it below
```

 
 
## Below examples are for aggregating and plotting:

```python
df.sort_values(by='MATH_GRADE', ascending=False)
```

```python
n_students = df.shape[0]

df['group'] = np.random.choice(['a','b','c'], n_students)
df
```

```python
# below is like:  SELECT min('math') FROM students GROUP BY group;

df.groupby('group', as_index=False)['MATH_GRADE'].min()
# this grabs the minimum of the math_grade field
```

```python
df.groupby('group', as_index=False)['MATH_GRADE'].max()
# this grabs the maximum of the math_grade field

# research what 'as_index=False' means....
```

```python
# this grabs items from several fields and aggregates them all together

# it looks at the READ_GRADE field by the three group (a,b,c), and groups it,
# returning the aggregated mean, min and max for the groups, specifically
# with regard to the reading grade.

df[['group', 'READ_GRADE']].groupby('group').agg([np.mean,'min','max'])
```

```python
# or we can grab one measure at a time

df[['group', 'READ_GRADE']].groupby('group').agg(['median'])
```

```python
# count
df[['group', 'READ_GRADE']].groupby('group').agg(['count'])
```
