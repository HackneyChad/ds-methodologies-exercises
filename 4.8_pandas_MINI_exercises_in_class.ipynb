10. Create a new column that contains the z-score for the shoe size.

```python
# these create the cts, mean, and stdev for the values in the shoe column
a_shoe_ct = students.shoe_sizes.count()
b_shoe_mean = students.shoe_sizes.mean()
c_shoe_std = students.shoe_sizes.std()

# then put them together in the formula to find z-score
z_shoe_score = (a_shoe_ct - b_shoe_mean) / c_shoe_std

# then create the new column in the data, naming it 'z_shoe_score' from the z-formula above,
# and setting that new field equal to the z-values
students['z_shoe_score'] = z_shoe_score
students

# can also just use 'shoe_size' without a "count" command, since it's in a data frame,
# and each series has the correct number of elements to make this work

# from yesterday's lesson, to find z-score  
# a = np.array([4, 10, 12, 23, -2, -1, 0, 0, 0, -6, 3, -7])
# b = a.mean()
# c = a.std()
# d = (a - b) / c
# d
```

## Aggregation & Plotting

1. Calculate the mean, median, min, and max for the shoe sizes and favorite numbers

```python
print('Shoe Size details:')
students.shoe_sizes.mean()
print(f'shoe_mean: {students.shoe_sizes.mean()}')

students.shoe_sizes.median()
print(f'shoe_median: {students.shoe_sizes.median()}')

students.shoe_sizes.min()
print(f'shoe_min: {students.shoe_sizes.min()}')

students.shoe_sizes.max()
print(f'shoe_max: {students.shoe_sizes.max()}\n')

print('Favorite Numbers details:')
students.fav_number.mean()
print(f'fav_no_mean: {students.fav_number.mean()}')

students.fav_number.median()
print(f'fav_no_median: {students.fav_number.median()}')

students.fav_number.min()
print(f'fav_no_min: {students.fav_number.min()}')

students.fav_number.max()
print(f'fav_no_max: {students.fav_number.max()}')
```
