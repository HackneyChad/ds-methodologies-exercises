### 4. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(12, 8))

x1 = list(range(1, 150))
y = [(math.sqrt(n)) for n in x1]

x2 = list(range(-50, 150))
z = [(n ** 3) for n in x2]

x2 = list(range(-50, 150))
a = [(math.tan(n)) for n in x2]

x1 = list(range(1, 150))
b = [(2 ** n) for n in x1]

# the first subplot
plt.subplot(221)
plt.plot(x1, y)
plt.title('x1 ~ y')

# the second subplot
plt.subplot(222)
plt.plot(x2, z)
plt.title('x2 ~ z')

# the third subplot
plt.subplot(223)
plt.plot(x2, a)
plt.title('x2 ~ a')

# the fourth subplot
plt.subplot(224)
plt.plot(x1, b)
plt.title('x1 ~ b')

plt.suptitle('Q4. - Subplots derived from Q3.')
plt.show()
```

#### 5. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points.  Be sure to include a legend.

```python
x1 = list(range(1, 150))
y = [(math.sqrt(n)) for n in x1]

x2 = list(range(-50, 150))
z = [(n ** 3) for n in x2]

x2 = list(range(-50, 150))
a = [(math.tan(n)) for n in x2]

x1 = list(range(1, 150))
b = [(2 ** n) for n in x1]

plt.plot(x1, y, c='red', label='y = √x')
plt.plot(x2, z, c='orange', label='y = x^3')
plt.xscale('linear') # the default
plt.yscale('log')
plt.plot(x2, a, c='yellow', label='y = tan(x)')
plt.plot(x1, b, c='green', label='y = 2^x')
plt.suptitle('Combining all plots together')
plt.legend()
plt.show()
```
