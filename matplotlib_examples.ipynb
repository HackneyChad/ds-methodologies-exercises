```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c='orange')
plt.plot(x, z, c='green')

plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 5),
            arrowprops={'facecolor': 'blue'})
plt.plot()
```

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')

plt.legend(loc='upper right')
plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')

plt.show()
```

```python
# histogram = frequency chart, like a bar chart, but actually a histogram
# think: frequency of items placed in 'bins'

x = [1, 1, 2, 3, 3, 3, 4, 4, 5]
# histograms only need an x, not a

plt.hist(x)
plt.show()
```

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6])
plt.show()
```

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.figure(figsize=(16, 6))
plt.suptitle(f'x = {x}')

# 131 below means one row and three columns
plt.subplot(131)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6])
plt.title('align="middle" (default)')
# align means alignment of frequency bars against the x axis markers... they align centered on the number
# on the x axis, or left or right of the number on the x axis.

plt.subplot(132)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='left')
plt.title('align="left"')

plt.subplot(133)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='right')
plt.title('align="right"')

plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i in range(1, 5):
    n = 10 ** i
    x1 = [randint(1, 6) for _ in range(n)]
    x2 = [randint(1, 6) for _ in range(n)]

    total = [x + y for x, y in zip(x1, x2)]

    plt.subplot(2, 2, i)
    plt.hist(total, bins=range(2, 14), align='left', color='white', edgecolor='black')
    plt.title(f'n = {n}')

plt.suptitle('Sum of 2 dice rolls')
plt.show()
```

```python
# Axis Scales
# We can change the scales of the axises to be, for example, logarithmic.

x = [n / 100 for n in range(-10000, 10000)]
y = list(x)

plt.plot(x, y)
plt.xscale('linear') # the default
plt.yscale('log')
plt.show()
```
